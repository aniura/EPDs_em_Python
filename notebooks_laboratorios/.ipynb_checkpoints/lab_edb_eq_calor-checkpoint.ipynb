## 1. Configurações iniciais.

```python
#Importamos a biblioteca SymPy para fazer cálculos simbólicos (ou seja, sem avaliar as expressões).
import sympy as sp

#Importamos algumas funções do módulo display da biblioteca IPython para obter uma impressão mais agradável.
from IPython.display import display, Math

#Importamos a biblioteca NumPy.
import numpy as np

#Importamos o módulo pyplot da biblioteca Matplotlib para fazer gráficos.
import matplotlib.pyplot as plt
```

**Exemplo 1**: O problema 

 $$\begin{array}{ll}
{\bf \rm ED} & \partial_t u=2\partial_x^2 u,\quad   0\leq x \leq \pi, t\geq 0; \\
{\bf \rm CF}  & u(0,t)=0 \quad u(\pi,t) =0, \quad  t\geq 0;\\
{\bf \rm CI} & u(x,0) = \text{sen}(2x),\quad  0 \leq x \leq \pi.   
\end{array}$$

tem solução $u_1(x,t) = e^{-8t} \text{sen}(2x)$. Isto pode ser verificado substituindo na equação, nas condições de fronteira e na condição inicial.

```python
#Definimos as variáveis simbólicas x e t
x, t = sp.symbols('x t')
#Definimos u_1 como expressão de x e t
u1 = sp.exp(-8*t)*sp.sin(2*x)
```

```python
#Verificamos a equação

#Calculamos a primeira derivada de u1 em relação a t
der_u1_t = sp.diff(u1, t, 1)
display(Math(f'\\partial_t u_1(x,t) = {der_u1_t}'))
#Calculamos a segunda derivada de u1 em relação a x
der_u1_2_x = sp.diff(u1, x, 2)
display(Math(f'\\partial^2_x u_1(x,t) = {der_u1_2_x}'))

#Substituímos na equação

display(Math(f'\\partial_t u_1(x,t) -2 \\partial^2_x u_1(x,t) = {der_u1_t -2 * der_u1_2_x}'))
```

```python
#Verificamos as condições de fronteira

display(Math(f'u_1(0,t) = {u1.subs(x, 0)}'))
display(Math(f'u_1(\\pi,t) = {u1.subs(x, sp.pi)}'))
```

```python
#Verificamos a condição inicial

display(Math(f'u_1(x,0) = {u1.subs(t, 0)}'))
```

```python
#Gráfico de u_1 como superfície

# Criar malha de pontos no quadrado [0,pi]x[0,1]
malha_x = np.arange(0, np.pi, 0.1)
malha_t = np.arange(0, 1, 0.1)

#X,Y contêm os pontos da malha quadrada
X, T = np.meshgrid(malha_x, malha_t)

#Valores da função u_1(x,y) = x, eles correspondem ao eixo z.
Z = np.exp(-8*T)*np.sin(2*X)

# Plotar a superfície
#subplot_kw={"projection": "3d"} indica que o gráfico é tridimensional
fig, ax = plt.subplots(figsize = (9,9), subplot_kw={"projection": "3d"})

ax.plot_surface(X, T, Z, vmin=Z.min() * 2, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('z')

plt.title(r'Gráfico de $u_1(x,t) = e^{-8t} sen(2x)$')

plt.show()
```

Podemos também fazer uma animação!

```python
#%matplotlib inline

#Módulo para fazer animações
import matplotlib.animation as animation

#Função para visualizar dados 
from IPython.display import HTML

plt.rcParams["animation.html"] = "jshtml"

def u_1(x, t):
   return np.exp(-8*t)*np.sin(2*x)
```

```python
#Parâmetros da figura
figsize = (6, 6)
colunas = 1
linhas = 1

#Número de instantes de tempo
N = 20

fig, ax = plt.subplots(linhas, colunas, figsize = figsize)

#Valores de x e t que vamos plotar
malha_x = np.linspace(0,np.pi,100)

#Elementos do primeiro quadro
ln, = ax.plot([], [])
title = ax.text(1.5,1.1, '')
    
#configurações iniciais do gráfico
def init():
    ax.set_xlim(-0.1, np.pi + 0.1)
    ax.set_ylim(-1.1, 1.2)
    return ln, title,

#Atualização do gráfico
def update(j):
    ln.set_color('tab:blue')
    ln.set_data(malha_x, u_1(malha_x,j/N))
    title.set_text(f't={j/N}')
    return ln, title,

#ani é uma instância da classe FuncAnimation
anim = animation.FuncAnimation(fig, update, frames=N, init_func=init, interval = 200)

#Não queremos mostrar a figura
plt.close()
#Mostrar a animação
anim
```

**Exemplo 2**: 

Podemos verificar também que a função $u_2(x,t) = e^{-18t} \text{sen}(3x)$ satisfaz o mesmo problema com $f(x) = \text{sen}(3x)$.

Consideremos $v(x,t) = u_1(x,t) - u_2(x,t) = e^{-8t} \text{sen}(2x) - e^{-18t} \text{sen}(3x)$. 

Sabemos que

+ Pelo Princípio de Superposição $v$  satisfaz a equação.
  
+ $v(0,t) = u_1(0,t) - u_2(0,t) =  0 - 0 = 0.$

+ $v(\pi,t) = u_1(\pi,t) - u_2(\pi,t) = 0 -  0 = 0.$

+ $v(x,0) = u_1(x,0) - u_2(x,0) = \text{sen}(2x) - \text{sen}(3x).$

Portanto, a solução do problema 

 $$\begin{array}{ll}
{\bf \rm ED} & \partial_t u=2\partial_x^2 u,\quad   0\leq x \leq \pi, t\geq 0; \\
{\bf \rm CF}  & u(0,t)=0 \quad u(\pi,t) =0, \quad  t\geq 0;\\
{\bf \rm CI} & u(x,0) = \text{sen}(2x) - \text{sen}(3x),\quad  0 \leq x \leq \pi.   
\end{array}$$

é $u_1(x,t) = e^{-8t} \text{sen}(2x) - e^{-18t} \text{sen}(3x)$.

**Exemplo 3** (constantes arbitrárias): 

Sejam $c_1$ e $c_2$ são constantes quaisquer e seja $v(x,t) = c_1 u_1(x,t) + c_2 u_2(x,t)$. Sabemos que

+ Pelo Princípio de Superposição $u$  satisfaz a equação.
  
+ $v(0,t) = c_1 u_1(0,t) + c_2 u_2(0,t) = c_1 \cdot 0 + c_2 \cdot 0 = 0.$

+ $v(\pi,t) = c_1 u_1(\pi,t) + c_2 u_2(\pi,t) = c_1 \cdot 0 + c_2 \cdot 0 = 0.$

+ $v(x,0) = c_1 u_1(x,0) + c_2 u_2(x,0) = c_1\text{sen}(2x) + c_2\text{sen}(3x).$

Portanto, a solução do problema 

 $$\begin{array}{ll}
{\bf \rm ED} & \partial_t u=2\partial_x^2 u,\quad   0\leq x \leq \pi, t\geq 0; \\
{\bf \rm CF}  & u(0,t)=0 \quad u(\pi,t) =0, \quad  t\geq 0;\\
{\bf \rm CI} & u(x,0) = c_1 \text{sen}(2x) + c_2 \text{sen}(3
x),\quad  0 \leq x \leq \pi.   
\end{array}$$

é $v(x,t) = c_1 e^{-8t} \text{sen}(2x) + c_2 e^{-18t} \text{sen}(3x)$.

**Generalizando:**

Para cada $n \in \mathbb{N}$, $u_n(x,t) = e^{-kn^2t} \text{sen}(nx)$ satisfaz

$$\begin{array}{ll}
{\bf \rm ED} & \partial_t u = k\partial_x^2 u,\quad   0\leq x \leq \pi, t\geq 0; \\
{\bf \rm CF}  & u(0,t)=0 \quad u(\pi,t) =0, \quad  t\geq 0;\\
{\bf \rm CI} & u(x,0) = \text{sen}(nx),\quad  0 \leq x \leq \pi.   
\end{array}$$

Portanto, a solução do problema

$$\begin{array}{ll}
{\bf \rm ED} & \partial_t u = k\partial_x^2 u,\quad   0\leq x \leq \pi, t\geq 0; \\
{\bf \rm CF}  & u(0,t)=0 \quad u(\pi,t) =0, \quad  t\geq 0;\\
{\bf \rm CI} & u(x,0) = b_1\text{sen}(x) + b_2\text{sen}(2x) + b_3\text{sen}(3x) + \cdots + b_N\text{sen}(Nx),\quad  0 \leq x \leq \pi
\end{array}$$

é

$$u(x,t) = b_1e^{-kt} \text{sen}(x) + b_2e^{-4kt}\text{sen}(2x) + b_3e^{-9kt}\text{sen}(3x) + \cdots + b_Ne^{-kN^2t}\text{sen}(Nx).$$
