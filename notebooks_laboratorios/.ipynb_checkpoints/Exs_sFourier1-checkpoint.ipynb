# Séries de Fourier de funções de período $2L$.

Para uma função $f: \mathbb{R} \rightarrow \mathbb{R}$ de período $2L$ tal que todas as integrais
 \begin{align}
  \label{coef_an}
  a_n&=\frac{1}{L}\int_{-L}^L f(x)\cos\left(\frac{n\pi}{L}x\right)dx,\quad n \geq 0;\\
  \label{coef_bn}
  b_n&=\frac{1}{L}\int_{-L}^L f(x)\text{sen}\left(\frac{n\pi}{L}x\right)dx,\quad n \geq 1. 
 \end{align}
estão bem definidas, a função
\begin{equation*}
 Sf(x)=\frac{a_0}{2}+\sum_{n=1}^{\infty}\left[a_n \cos\left(\frac{n\pi}{L}x\right)+
 b_n\text{sen}\left(\frac{n\pi}{L}x\right)\right]
\end{equation*}
chama-se de **série de Fourier** de $f$ (caso ela esteja bem definida). 

Os valores $a_n$ e $b_n$ são chamados de **coeficientes de Fourier** de $f$ e a convergência da série deve ser entendida no sentido da convergência das somas parciais de Fourier
\begin{equation}
 S_Nf(x)=\frac{a_0}{2}+\sum_{n=1}^N\left[a_n \cos\left(\frac{n\pi}{L}x\right)+
 \text{sen}\left(\frac{n\pi}{L}x\right)\right],
\end{equation}
quando $N \rightarrow \infty$. 

Nesta notebook vamos ilustrar o comportamento dessas somas parciais, da série e dos coeficientes de Fourier. 

```python
#NumPy é a principal biblioteca para cálculos científicos de Python. Ela disponibiliza o objeto multidimensional 
#arranjo, objetos derivados, e um conjuntos de rotinas para operações rápidas sobre arranjos.
import numpy as np

#Matplotlib é uma biblioteca abrangente para criar visualizações estáticas, animadas e interativas em Python
import matplotlib.pyplot as plt

#%matplotlib inline
#SymPy é uma biblioteca para cálculos simbólicos como integração exata, derivação, 
#resolução de equações, expansão em série de Taylor, etc
import sympy as sp 

#Importamos algumas funções do módulo display da biblioteca IPython para obter uma impressão mais agradável.
from IPython.display import display, Math

# Fonte a ser usada nas figuras geradas pela Matplotlib.
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 12
```

## Exemplo 1:

$$f(x) = \begin{cases}
          0, & -1 \leq x < 0;\\
          1, & 0 \leq x < 1;\\
          \text{ periódica de período }2\pi.
         \end{cases}$$
Graficamente:         

```python
L = np.pi #Metade de um período
M = 100 # Quantidade de pontos numa malha para plotar

# Tamanho da figura e janelas
figsize = (9, 6)
colunas = 1
linhas = 1
fig, axs = plt.subplots(linhas, colunas, figsize=figsize)
fig.suptitle('Gráfico de f')
axs.set_ylabel('f(x)')
axs.set_xlabel('x')
axs.set_xlim((-3*L-0.1,3*L+0.1))
axs.set_ylim((-0.1,1.1))

#Intervalos onde a função vale 0
axs.plot(np.linspace(-3*L, -2*L, num=M + 1), np.zeros(M + 1),'-',color = 'blue')
axs.plot(np.linspace(-L, 0, num=M + 1), np.zeros(M + 1),'-',color = 'blue')
axs.plot(np.linspace(L, 2*L, num=M + 1), np.zeros(M + 1),'-',color = 'blue')

#Intervalos onde a função vale 1
axs.plot(np.linspace(-2*L, -L, num=M + 1), np.ones(M + 1),'-',color = 'blue')
axs.plot(np.linspace(0, L, num=M + 1), np.ones(M + 1),'-',color = 'blue')
axs.plot(np.linspace(2*L, 3*L, num=M + 1), np.ones(M + 1),'-',color = 'blue')

#Pontos onde a função tem saltos
axs.plot([-2*L, 0, 2*L], [0, 0, 0],'bo', markersize = 8, fillstyle = 'none')
axs.plot([-L, L, 3*L], [1, 1, 1],'bo', markersize = 8, fillstyle = 'none')
axs.grid(True)
#axs.legend(loc='best');
```

### a) Cálculo dos coeficientes de Fourier

```python
#coeficientes de Fourier de f
x = sp.Symbol('x')#x será considerado como símbolo
n = sp.Symbol('n', integer = True, positive=True)
#L = var('L', nonzero=True)
#coefte a0: expressão simbólica
a0 = sp.integrate(1,(x,0,L))/L
#coeftes an: expressão simbólica
#an = lambda n:sp.trigsimp((1/L)*sp.integrate(sp.cos(sp.pi*n*x/L),(x,0,L)))
an = sp.trigsimp(sp.integrate(sp.cos(sp.pi*n*x/L),(x,0,L))/L)
#coeftes bn: expressão simbólica
bn = sp.trigsimp(sp.integrate(sp.sin(sp.pi*n*x/L),(x,0,L))/L)
#Mostrar os resultados
#a0
display(Math(f'a_0 = {sp.latex(a0)}'))
#an
display(Math(f'a_n = {sp.latex(an)}'))
#bn
display(Math(f'b_n = {sp.latex(bn)}'))
```

### b) Somas parciais de Fourier

```python
#Sequência de coeficientes an
a = [an.evalf(subs={n: j}) for j in range(1,50)]
#Sequência de coeficientes bn
b = [bn.evalf(subs={n: j}) for j in range(1,50)]
```

### c) Gráficos dos coeficientes de Fourier

```python
# Tamanho da figura e janelas
figsize = (12, 6)
colunas = 2
linhas = 1
fig_coeftes, axs = plt.subplots(linhas, colunas, figsize=figsize, constrained_layout=True)
fig_coeftes.suptitle('Gráficos dos coeficientes de Fourier')
N = 30

#Coeficientes a_n
axs[0].set_title(r'Coeficientes $a_n$')
axs[0].set_ylabel(r'$a_n$')
axs[0].set_xlabel('n')
axs[0].set_xlim((-0.1,N + 0.1))
axs[0].set_ylim((-1.2,1.2))
axs[0].plot(np.linspace(1, N, num = N), a[:N],'o', color = 'red', markersize = 4)
axs[0].plot([0], [a0],'o', color = 'red', markersize = 6)
axs[0].grid(True)

#Coeficientes b_n
axs[1].set_title(r'Coeficientes $b_n$')
axs[1].set_ylabel(r'$b_n$')
axs[1].set_xlabel('n')
axs[1].set_xlim((-0.1,N + 0.1))
axs[1].set_ylim((-1.2,1.2))
axs[1].plot(np.linspace(1, N, num = N), b[:N],'o', color = 'blue', markersize = 4)
axs[1].grid(True);
```

### d) Gráficos de f junto com a série e com as somas parciais de Fourier

```python
#Função que calcula uma soma parcial de Fourier, como função de Python
def S_N(xv,N):
    #Soma parcial de Fourier da solução
    result = a0/2 + sum([a[j-1]*np.cos(j*np.pi*xv/L) for j in range(1,N+1)]) + \
    sum([b[j-1]*np.sin(j*np.pi*xv/L) for j in range(1,N+1)])
    #soma parcial que pode ser avaliada no Python 
    return result
```

```python
# Tamanho da figura e janelas
figsize = (12, 6)
colunas = 2
linhas = 1
fig, axs = plt.subplots(linhas, colunas, figsize=figsize, constrained_layout=True)
fig.suptitle(r'Gráficos de $f$, $Sf(x)$ e $S_N f$')

#Função $f$ junto com as somas parciais
axs[0].set_title(r'$f$ e $S_N f$')
axs[0].set_ylabel('f(x)')
axs[0].set_xlabel('x')
axs[0].set_xlim((-3*L-0.1,3*L+0.1))
axs[0].set_ylim((-0.2,1.2))

#Intervalos onde a função vale 0
axs[0].plot(np.linspace(-3*L, -2*L, num=M + 1), np.zeros(M + 1),'-',color = 'blue')
axs[0].plot(np.linspace(-L, 0, num=M + 1), np.zeros(M + 1),'-',color = 'blue')
axs[0].plot(np.linspace(L, 2*L, num=M + 1), np.zeros(M + 1),'-',color = 'blue')

#Intervalos onde a função vale 1
axs[0].plot(np.linspace(-2*L, -L, num=M + 1), np.ones(M + 1),'-',color = 'blue')
axs[0].plot(np.linspace(0, L, num=M + 1), np.ones(M + 1),'-',color = 'blue')
axs[0].plot(np.linspace(2*L, 3*L, num=M + 1), np.ones(M + 1),'-',color = 'blue')

#Pontos onde a função tem saltos
axs[0].plot([-2*L, 0, 2*L], [0, 0, 0],'bo', markersize = 6, fillstyle = 'none')
axs[0].plot([-L, L, 3*L], [1, 1, 1],'bo', markersize = 6, fillstyle = 'none')

#Plotar somas parciais de Fourier
malha_x = np.linspace(-3*L, 3*L, num=10*M + 1)
axs[0].plot(malha_x, S_N(malha_x,5),'-',color = 'magenta',linewidth=1, label = 'N=5')
axs[0].plot(malha_x, S_N(malha_x,10),'-',color = 'orange',linewidth=1, label = 'N=10')
axs[0].plot(malha_x, S_N(malha_x,15),'-',color = 'green',linewidth=1, label = 'N=15')
axs[0].plot(malha_x, S_N(malha_x,15),'-',color = 'tab:blue',linewidth=1, label = 'N=30')


axs[0].grid(True)
axs[0].legend(loc='lower right')

#Série de Fourier junto com as somas parciais
axs[1].set_title(r'$Sf$ e $S_N f$')
axs[1].set_ylabel('Sf(x)')
axs[1].set_xlabel('x')
axs[1].set_xlim((-3*L-0.1,3*L+0.1))
axs[1].set_ylim((-0.2,1.2))

#Intervalos onde a função vale 0
axs[1].plot(np.linspace(-3*L, -2*L, num=M + 1), np.zeros(M + 1),'-',color = 'red')
axs[1].plot(np.linspace(-L, 0, num=M + 1), np.zeros(M + 1),'-',color = 'red')
axs[1].plot(np.linspace(L, 2*L, num=M + 1), np.zeros(M + 1),'-',color = 'red')

#Intervalos onde a função vale 1
axs[1].plot(np.linspace(-2*L, -L, num=M + 1), np.ones(M + 1),'-',color = 'red')
axs[1].plot(np.linspace(0, L, num=M + 1), np.ones(M + 1),'-',color = 'red')
axs[1].plot(np.linspace(2*L, 3*L, num=M + 1), np.ones(M + 1),'-',color = 'red')

#Pontos onde a série tem saltos
axs[1].plot(np.linspace(-3*L, 3*L, num = 7), 0.5*np.ones(7),'o', color = 'red', markersize = 6)
axs[1].plot(np.linspace(-3*L, 2*L, num = 6), np.zeros(6),'o', color = 'red', markersize = 6, fillstyle = 'none')
axs[1].plot(np.linspace(-2*L, 3*L, num = 6), np.ones(6),'o', color = 'red', markersize = 6, fillstyle = 'none')

#Plotar somas parciais de Fourier
malha_x = np.linspace(-3*L, 3*L, num=10*M + 1)
axs[1].plot(malha_x, S_N(malha_x,5),'-',color = 'magenta',linewidth=1, label = 'N=5')
axs[1].plot(malha_x, S_N(malha_x,10),'-',color = 'orange',linewidth=1, label = 'N=10')
axs[1].plot(malha_x, S_N(malha_x,15),'-',color = 'green',linewidth=1, label = 'N=15')
axs[1].plot(malha_x, S_N(malha_x,15),'-',color = 'tab:blue',linewidth=1, label = 'N=30')

axs[1].grid(True)
axs[1].legend(loc='lower right');
```
